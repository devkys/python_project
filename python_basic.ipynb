```python
import ipywidgets as widgets
from IPython.display import display

slider = widgets.IntSlider(min=0, max=200, step=5, description='intslider')
boolean = widgets.Checkbox(value=False, description='checkbox')
txt = widgets.Text(vlaue='hi', placeholder='type here', description='Text')
btn=widgets.Button(description='Button')

def btn_click(b):
    print(slider.value, '\n', boolean.value, '\n', txt.value)
btn.on_click(btn_click)
display(slider, boolean, txt, btn)
    
```

```python
import ipywidgets as widgets
from IPython.display import display

name = widgets.Text(vlaue='', placeholder='input your name', description='Name')
birth = widgets.Text(vlaue='', placeholder='input your birth', description='Birth')
man_boolean = widgets.Checkbox(value=False, description='Man')
woman_boolean = widgets.Checkbox(value=False, description='Woman')
slider = widgets.IntSlider(min=0, max=200, step=5, description='Age')

def btn_click(b):
    print(name.value, '\n',birth.value, '\n',man_boolean.value, '\n', woman_boolean.value, '\n', slider.value)
btn.on_click(btn_click)
display(name, birth, man_boolean, woman_boolean, slider, btn)
```
